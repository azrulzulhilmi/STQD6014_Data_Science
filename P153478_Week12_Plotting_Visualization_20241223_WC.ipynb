### Histogram


*   **`bins`**: Bins are the intervals into which the data is divided for counting and display in the histogram.
    - Having 20 bins means the data range will be split into 20 intervals
*   **`alpha`**: controls the transparency of the bars.
    - A value of 1.0 would be completely opaque, while 0.0 would be completely transparent.

### **Place all the subplots into one panel**

1.   **plt.figure()** - create a new figure
2.   **add_subplot** - to create one or more subplots
3.   **Subplots** are numbered sequentially **starting from 1, going from left to right and top to bottom.**
    - a **(2, 2, 1)** argument means we're adding the first subplot (top-left corner) in the 2x2 grid.


```python
# Create one or more subplots
# 2 x 2; 4 plots in total
# Plot axis
rng = np.random.RandomState(12345)
fig = plt.figure(dpi=300) #dpi: dot per image
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax1.hist(rng.standard_normal(100), bins=20, color='black', alpha=0.3);
ax2.scatter(np.arange(30), np.arange(30) + 3 * rng.standard_normal(30));
ax3.plot(rng.standard_normal(50).cumsum(), color='black', linestyle='dashed');

# To save the figures
plt.savefig('/content/drive/MyDrive/STQD6014/combi3.pdf')
```

### **Adjusting the spacing around subplots**

```python
# An example that shrink the spacing of the subplots to zero
# Change the value of wspace and hspace
rng = np.random.RandomState(12345)
fig,axes = plt.subplots(2, 2, sharex=True, sharey=False)

# Using a nested loop to populate 2x2 grid
for i in range(2):
    for j in range(2):
        axes[i, j].hist(rng.standard_normal(500), bins=50,
                        color='black', alpha=0.7)

# Adjusting the spacing
fig.subplots_adjust(wspace=0.3, hspace=0.3)
```

### ***Setting the title, axis labels, ticks, and tick labels***

```python
# An example
rng = np.random.RandomState(12345)
fig, ax = plt.subplots()
ax.plot(rng.standard_normal(1000).cumsum());

# Changex x-axis ticks annotation
ax.set_xticks([0, 250, 500, 750, 1000]);

# Label the x-axis ticks with names
ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'],
                   rotation=25, fontsize=10)
ax.set_xlabel('Stages')
ax.set_ylabel('Values')
ax.set_title('My first matplotlib plot');
```

### **Annotations and Drawing on a Subplot**

```python
# The SPX data
from datetime import datetime

fig, ax = plt.subplots()

data = pd.read_csv('https://bit.ly/3iofIY8', index_col=0, parse_dates=True)
#parse_date: This argument tells Pandas to attempt to parse (interpret) any columns
#that look like dates into actual datetime objects.
# index_col=0 : sep=""?
spx = data['SPX']
spx
spx.plot(ax=ax, color='blue');

crisis_data = [
    (datetime(2007, 10, 11), 'Peak of bull market'),
    (datetime(2008, 3, 12), 'Bear Stearns Fails'),
    (datetime(2008, 9, 15), 'Lehman Bankruptcy')
] #label the peak

for date, label in crisis_data:
    ax.annotate(label, xy=(date, spx.asof(date) + 75),
                xytext=(date, spx.asof(date) + 225),
                arrowprops=dict(facecolor='black', headwidth=4, width=2,
                                headlength=4),
                horizontalalignment='left', verticalalignment='top')

# Zoom in on 2007-2010
ax.set_xlim(['1/1/2007', '1/1/2011']) #choose date by using set
ax.set_ylim([600, 1800]) #choose limit for y
ax.set_title('Important dates in the 2008-2009 financial crisis');
```

### **Adding shapes on the same plot**

```python
# Adding shapes to the same plot #exam
fig, ax = plt.subplots()

rect = plt.Rectangle((0.2, 0.75), 0.4, 0.15, color='black', alpha=0.3)
circ = plt.Circle((0.7, 0.2), 0.15, color='blue', alpha=0.3)
pgon = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]],
                   color='green', alpha=0.5)
ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon);
```

```python
# Challenge: plot a pentagon, ellipse, square figures in the exact same figure
fig, ax = plt.subplots()

# Pentagon
  # Center, sides, radius
pentagon=patches.RegularPolygon((0.2,0.7),5,radius=0.15,color='red')
ax.add_patch(pentagon)

# Ellipse
  # Center, width, height
ellipse=patches.Ellipse((0.7,0.7),0.2,0.1,color='navy')
ax.add_patch(ellipse)

# Square
  # Lower-left, width, height
square=patches.Rectangle((0.3,0.1),0.2,0.2,color='green',alpha=0.5)
ax.add_patch(square);

plt.show()  # Display the figure
```

```python
# Adding shapes to a plot
import matplotlib.patches as mpatches
fig, ax = plt.subplots()

rect = plt.Rectangle((0.2, 0.75), 0.4, 0.15, color='black', alpha=0.3)
circ = plt.Circle((0.7, 0.2), 0.15, color='blue', alpha=0.3)
pgon = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]],
                   color='green', alpha=0.5)
sqr = plt.Rectangle((0.3, 0.1), 0.2, 0.2, color='black', alpha=0.3)

# Add ellipse
ellipse = mpatches.Ellipse((0.5,0.5), 0.2, 0.1)

# Add regular hex
hex = mpatches.RegularPolygon((0.7,0.8), 6, radius=0.1)

ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)
ax.add_patch(sqr)
ax.add_patch(ellipse)
ax.add_patch(hex);
#plt.savefig('/content/drive/MyDrive/STQD6014/shape.pdf',dpi=400)
```
